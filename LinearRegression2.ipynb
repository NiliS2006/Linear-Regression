**Math behind Logistic Regression**

---



**y=mx+b**
------
where:
y=estimated dependent value
b=bias
m=regression coefficient or slope
x=value of independent variable

**Loss Function**


---

mean squared error = (sigma(y1-y2)^2)/n
y1 and y2 are true and predicted values

**Optimization Algorithm**

---

Gradient Descent


---

slope = theta and constant = theta.
calculating partial derivatives differentiating loss function wrt parameters like m and b, later updating the parameter values. each step minimizes loss function and reduces difference between true and predicted values.


```python
#Defining the class
class LinearRegression:
    def __init__(self, x , y):
        self.data = x
        self.label = y
        self.m = 0 #slope
        self.b = 0 #intercept
        self.n = len(x)

    def fit(self , epochs , lr):
    # lr is the learning rate
    #epochs is the number of interactions over training set
        #Implementing Gradient Descent
        for i in range(epochs):
            y_pred = self.m * self.data + self.b

            #Calculating derivatives w.r.t Parameters
            D_m = (-2/self.n)*sum(self.data * (self.label - y_pred))
            D_b = (-1/self.n)*sum(self.label-y_pred)

            #Updating Parameters # lr = 0.01
            self.m = self.m - lr * D_m
            self.c = self.b - lr * D_b

    def predict(self , inp):
        y_pred = self.m * inp + self.b
        return y_pred
```

```python
#importing Matplotlib for plotting
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#Loding the data
df = pd.read_excel('/content/lineardataset.xlsx')

#Preparing the data
x = np.array(df.iloc[:,0])
y = np.array(df.iloc[:,1])

#Creating the class object
regressor = LinearRegression(x,y)

#Training the model with .fit method
regressor.fit(1000 , 0.0001) # epochs-1000 , learning_rate - 0.0001

#Prediciting the values
y_pred = regressor.predict(x)

#Plotting the results
plt.figure(figsize = (10,6))
plt.scatter(x,y , color = 'red')
plt.plot(x , y_pred , color = 'g' , lw = 1)
plt.xlabel('x' , size = 20)
plt.ylabel('y', size = 20)
plt.show()
```
